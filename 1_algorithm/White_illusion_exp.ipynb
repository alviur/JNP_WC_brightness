## Parameters

```julia
sigmaMu = [1,2,3,5,10,20,30]
sigmaW = [1,2,5,10,20]
lamda = [0.2,0.5,0.8,1.2]
```

```julia
images = ["white0.01.png","white0.03.png","white0.05.png","white0.07.png","white0.09.png","white0.12.png"]
savePath = "/home/alexander/Desktop/JNP_italians/3_Results/whiteWC_cortical/"
for m = 1:size(sigmaMu,1)
    
    for w = 1:size(sigmaW,1)
        for l = 1:size(lamda,1)
            
            for i = 1:size(images,1)
                print(m,w,l,i)

                println(images[i])
                img = Gray.(load(string(path,"/",images[i])))
                
                ## LHE planar
                #img_lhe_3d = lhe(img, sigmaMu[m], sigmaW[w], lamda[l], algo_type = :planar)[1]
                img_lhe_3d = wc(img, sigmaMu[m], sigmaW[w], lamda[l], max_iter = 100, threshold = 0.01, m=200, bw = 4, algo_type = :cortical)[1]
                save(string(savePath,images[i],string(m),"_",string(w),"_",string(l),".jld"), "data", img_lhe_3d)

            end
        end
    end
end
```
