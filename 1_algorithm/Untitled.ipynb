## White lhe Planar

```python
path = '/home/alexander/Desktop/JNP_italians/3_Results/param_variation/White/whiteCortical/'
xpoints = [(64,143),(64,153),(66,128),(64,164),(46,172),(61,133)]
images = ["white0.01.png","white0.03.png","white0.05.png","white0.07.png","white0.09.png","white0.12.png"]
sigmaMu = [1,2,3,5,10,20,30]
sigmaW = [1,2,5,10,20]
lamda = [0.2,0.5,0.8,1.2]



contImages = 0
response = np.zeros((len(images),len(sigmaMu)*len(sigmaW)*len(lamda)))

for m in range(1,len(sigmaMu)+1):
    for w in range (1,len(sigmaW)+1):
        for l in range(1,len(lamda)+1):
            for i in range(len(images)):
                
                

                print(images[i])              
                ## LHE planar               
                f = h5py.File(path+images[i]+str(m)+"_"+str(w)+"_"+str(l)+'.jld', "r")
                img = f["data"].value.astype(float).T
                
                left = xpoints[0][0]
                right = xpoints[0][1]
                
                response[i,contImages] = img[100,right]-img[100,left]
                
                
                print(response[i,contImages],img[100,right]-img[100,left])
            contImages += 1

```

```python
response[response>0]=1
response[response<=0]=0
a = np.sum(response,axis=0)
plt.imshow(response.T)

plt.show()
a
```
